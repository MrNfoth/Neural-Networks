# Результаты работы рецептивных полей

Ниже приведены визуализации откликов трёх типов клеток: **Ganglion**, **Simple** и **Complex**. Все изображения увеличены для читаемости.

---

## Ganglion Cell

### 1. Карта откликов на точечный стимул
- **Файл:** `ganglion_point.png`
- **Описание:** показывает локальный отклик ганглиозной клетки на каждый пиксель тестового поля.
- **Оси:** 
  - X и Y – координаты пикселей на тестовом поле.
  - Цвет (яркость) – величина отклика клетки на этот пиксель.
- Яркие точки показывают пиксели, где клетка реагирует сильнее.

<img src="aoutput/ganglion_point.png" width="600">

---

### 2. Нормализованная карта точечных откликов
- **Файл:** `ganglion_point_norm.png`
- **Описание:** значения нормализованы для удобной визуализации, подчеркиваются зоны максимальной реакции.
- **Оси:** 
  - X и Y – координаты пикселей поля.
  - Цвет – нормализованная величина отклика (0–255).
- Используется для оценки формы рецептивного поля.

<img src="aoutput/ganglion_point_norm.png" width="600">

---

### 3. Ориентационная селективность
- **Файл:** `ganglion_rotate.png`
- **Описание:** график отклика при вращении линии на разные углы (0–360°).
- **Оси:** 
  - X – угол линии (градусы)
  - Y – величина отклика клетки
- Пики графика показывают предпочтительные направления для данной клетки.

<img src="aoutput/ganglion_rotate.png" width="600">

---

### 4. Селективность по смещению линии
- **Файл:** `ganglion_shift.png`
- **Описание:** измеряется отклик при сдвиге вертикальной линии.
- **Оси:** 
  - X – смещение линии относительно центра (пиксели)
  - Y – величина отклика
- Показывает пространственную чувствительность.

<img src="aoutput/ganglion_shift.png" width="600">

---

### 5. Селективность по размеру (радиус круга)
- **Файл:** `ganglion_circle.png`
- **Описание:** график отклика при изменении радиуса заполненного круга.
- **Оси:** 
  - X – радиус круга (пиксели)
  - Y – величина отклика
- Позволяет определить оптимальный размер стимулирующей области для клетки.

<img src="aoutput/ganglion_circle.png" width="600">
